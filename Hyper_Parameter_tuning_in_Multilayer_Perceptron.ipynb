```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

                        SENARIO-1  SENARIO-2  SENARIO-3  SENARIO-4

NO OF EPOCH             24         25         21         23
VALIDATION SPLIT        0.3        0.2        0.1        0.2
OPTIMIZER               adagrad    adam       adadelta   adam
LEARNING RATE           1          1          1          1

```python
X_train = X_train/255
X_test  = X_test/255
```

```python
model1 = Sequential()
model1.add(Flatten (input_shape=(28,28)))
model1.add(Dense(128,activation='sigmoid'))
model1.add(Dense(32,activation='sigmoid'))
model1.add(Dense(10,activation='softmax'))

model1.compile(loss='sparse_categorical_crossentropy',optimizer='Adagrad',metrics=['accuracy'])
history1 = model1.fit(X_train,y_train,epochs=24,validation_split=0.3)

y_prob1 = model1.predict(X_test)
y_pred1 = y_prob1.argmax(axis=1)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```

```python
model2 = Sequential()
model2.add(Flatten (input_shape=(28,28)))
model2.add(Dense(128,activation='relu'))
model2.add(Dense(32,activation='relu'))
model2.add(Dense(10,activation='softmax'))

model2.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
history2 = model2.fit(X_train,y_train,epochs=25,validation_split=0.2)

y_prob2 = model2.predict(X_test)
y_pred2 = y_prob2.argmax(axis=1)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred2)
```

```python
model3 = Sequential()
model3.add(Flatten (input_shape=(28,28)))
model3.add(Dense(128,activation='tanh'))
model3.add(Dense(32,activation='tanh'))
model3.add(Dense(10,activation='softmax'))

model3.compile(loss='sparse_categorical_crossentropy',optimizer='Adadelta',metrics=['accuracy'])
history3 = model3.fit(X_train,y_train,epochs=21,validation_split=0.1)

y_prob3 = model3.predict(X_test)
y_pred3 = y_prob3.argmax(axis=1)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred3)
```

```python
model4 = Sequential()
model4.add(Flatten (input_shape=(28,28)))
model4.add(Dense(128,activation='sigmoid'))
model4.add(Dense(32,activation='sigmoid'))
model4.add(Dense(10,activation='softmax'))

model4.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
history4 = model4.fit(X_train,y_train,epochs=23,validation_split=0.2)

y_prob4 = model4.predict(X_test)
y_pred4 = y_prob4.argmax(axis=1)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred4)
```

```python
[plt.plot(history.history["loss"]) for history in [history1, history2, history3, history4]]
```

```python
[plt.plot(history.history["val_loss"]) for history in [history1, history2, history3, history4]]
```

```python
[plt.plot(history.history["val_accuracy"]) for history in [history1, history2, history3, history4]]
```

```python
for i, y_pred in enumerate([y_pred1, y_pred2, y_pred3, y_pred4], start=1):
    print("SCENARIO", i, ":", accuracy_score(y_test, y_pred) * 100, "%")
```
